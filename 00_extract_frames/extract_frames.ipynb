```python
import cv2
from pathlib import Path
import traceback
import matplotlib.pyplot as plt
from math import floor
```

```python
# define the path to the video
path_video = Path("/Users/johanneseder711/Documents/Python/01_active/Keepercam/Videos/221206_clip.MOV")
```

```python
# Opens the Video file
if path_video.exists():
    try:
        cap = cv2.VideoCapture(str(path_video))
    except:
        print(f'Something went wrong when loading your video: {traceback.print_exc()}')
else:
    print('Could not find the video you intended to load!')
```

```python
# get the number of frames per second
fps = cap.get(cv2.CAP_PROP_FPS)
print(f"The Video has approx. {int(round(fps,0))} frames per second.")
```

```python
# get the total number of frames in the video
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print(f"The Video has a total number of {frame_count} frames.")
```

```python
duration_in_sec = frame_count/fps
print(f"The Video takes {floor(duration_in_sec/60)} minutes and {int(round(duration_in_sec%60,0))} seconds.")
```

```python
# Capture frame-by-frame
ret, frame = cap.read()
```

```python
if ret: 
    plt.figure(figsize=(10,7))
    plt.imshow(frame)
    plt.axis(False)
    plt.show()
else:
    print('Could not read image!')
```

# Save image/frame every second

```python
# Opens the Video file
cap= cv2.VideoCapture(str(path_video))
framecounter = 1
second_frame = 1
while second_frame < 10:
    # read the current frame
    ret, frame = cap.read()
    if(framecounter%int(fps) == 0):
        if ret == False:
            break
        cv2.imwrite('frame_at_second_'+str(second_frame)+'.jpg',frame)
        second_frame+=1
    framecounter+=1
 
cap.release()
cv2.destroyAllWindows()
```
